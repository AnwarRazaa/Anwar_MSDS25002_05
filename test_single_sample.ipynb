# Test Single Sample
Loads the trained diffusion model checkpoint and generates new animal images by running the reverse (denoising) process starting from pure Gaussian noise.

```python
import torch
import matplotlib.pyplot as plt

from MSDS25002_05 import GaussianDiffusion, unnormalize
from MSDS25002_05_model import DenoiseUNet
```

```python
CHECKPOINT_PATH = "saved_models/diffusion_model.pt"

checkpoint = torch.load(CHECKPOINT_PATH, map_location="cpu")
img_size = checkpoint["img_size"]
timesteps = checkpoint["timesteps"]

model = DenoiseUNet(image_channels=3)
model.load_state_dict(checkpoint["model_state_dict"])
model.eval()

diffusion = GaussianDiffusion(timesteps=timesteps, device="cpu")
print(f"Loaded model trained at img_size={img_size}, timesteps={timesteps}")
```

## Generate a single image from pure noise

```python
with torch.no_grad():
    sample = diffusion.sample(model, image_size=img_size, batch_size=1)

plt.figure(figsize=(3, 3))
plt.imshow(unnormalize(sample[0]).permute(1, 2, 0).numpy())
plt.axis("off")
plt.title("Generated sample")
plt.show()
```

## Generate a small batch of samples

```python
with torch.no_grad():
    samples = diffusion.sample(model, image_size=img_size, batch_size=4)

fig, axes = plt.subplots(1, 4, figsize=(10, 3))
for ax, img in zip(axes, samples):
    ax.imshow(unnormalize(img).permute(1, 2, 0).numpy())
    ax.axis("off")
plt.tight_layout()
plt.show()
```
